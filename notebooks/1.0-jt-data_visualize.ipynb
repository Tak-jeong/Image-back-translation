```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file_path = "/mnt/nas65/Dataset/ImageNet1K/ILSVRC/Data/CLS-LOC/train_csv"
class_list = ['n02085936','n02134084','n04536866','n07745940','n04552348','n04548280','n07734744','n02783161','n02676566','n01910747',]

cls = class_list[9] # set class 

data_frame = pd.read_csv(csv_file_path + '/' + cls + ".csv", sep=r",(?:(?!\s)+(?!')+(?!$))", engine='python')
```

```python
real_image_path = data_frame.iloc[0,0]
real_image_path = real_image_path.replace("/home/datasets", "/mnt/nas65/Dataset/")
real_caption = data_frame.iloc[0,1]
diff_image_num = real_image_path.split("_")[-1].split(".")[0] # 0001
diff_image_path = os.path.dirname(real_image_path).replace("train", "train_diff") + "/" + diff_image_num + "_0.png"
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))  # Create a figure and a set of subplots

# Display the first image
axs[0].imshow(plt.imread(real_image_path))
axs[0].set_title('Real Image')
axs[0].axis('off')  # To hide x and y axis values

# Display the second image
axs[1].imshow(plt.imread(diff_image_path))
axs[1].set_title('Diff Image')
axs[1].axis('off')  # To hide x and y axis values

plt.suptitle(f"Class name: {cls}\nReal caption: {real_caption}", y=1.1)

plt.show()
```
